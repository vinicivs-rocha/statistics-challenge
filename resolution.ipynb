# Inicialização de dependências

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

# Processando a média

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])
```

```python
mean = df_faturamento['valor'].mean()
print(f'Média do faturamento: {mean}')
```

# Plotando gráficos

```python
df_faturamento = df_faturamento.sort_values(by='data_ref')
```

```python
ax = df_faturamento.plot.bar(x='data_ref', y='valor', title='Faturamento por mês')
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda y, p: format(int(y), ',')))
ax.set_xticklabels(df_faturamento['data_ref'].dt.strftime('%Y-%m'), rotation=45)
```

```python
ax = df_faturamento.plot.line(x='data_ref', y='valor', title='Faturamento por mês')
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda y, p: format(int(y), ',')))
ax.set_xticks(df_faturamento['data_ref'])
ax.set_xticklabels(df_faturamento['data_ref'].dt.strftime('%Y-%m'), rotation=80)
```
